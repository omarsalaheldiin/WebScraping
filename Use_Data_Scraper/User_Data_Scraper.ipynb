```python
import requests
import pandas as pd
```

```python
# Fetch the user data
res = requests.get('https://jsonplaceholder.typicode.com/users')
data = res.json()
```

```python
# Create a DataFrame
df = pd.DataFrame(columns=['Name', 'Username', 'Email', 'Street', 'Suite', 'City', 'Zip Code', 'Geo Latitude', 'Geo Longitude'])
```

```python
# Parse the user data
for idx, user in enumerate(data):
    address = user.get("address", {})
    geo = address.get("geo", {})
    
    df.loc[idx, 'Name'] = user.get("name")
    df.loc[idx, 'Username'] = user.get("username")
    df.loc[idx, 'Email'] = user.get("email")
    df.loc[idx, 'Street'] = address.get("street")
    df.loc[idx, 'Suite'] = address.get("suite")
    df.loc[idx, 'City'] = address.get("city")
    df.loc[idx, 'Zip Code'] = address.get("zipcode")
    df.loc[idx, 'Geo Latitude'] = geo.get("lat")
    df.loc[idx, 'Geo Longitude'] = geo.get("lng")
    
df
```

```python
# Save to CSV
df.to_csv('users_data.csv', index=False)
```
