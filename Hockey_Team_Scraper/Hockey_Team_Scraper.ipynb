```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Fetch the hockey teams information
res = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=1')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# Create a DataFrame
df = pd.DataFrame(columns=['Team Name', 'Year', 'Wins', 'Losses', 'Win %', 'Goals For (GF)', 'Goals Against (GA)'])
```

```python
# Parse the teams data
for idx, team in enumerate(soup.find_all('tr')):
    if idx == 0:
        continue
    df.loc[idx, 'Team Name'] = team.find_all('td')[0].text.strip()
    df.loc[idx, 'Year'] = team.find_all('td')[1].text.strip()
    df.loc[idx, 'Wins'] = team.find_all('td')[2].text.strip()
    df.loc[idx, 'Losses'] = team.find_all('td')[3].text.strip()
    df.loc[idx, 'Win %'] = team.find_all('td')[5].text.strip()
    df.loc[idx, 'Goals For (GF)'] = team.find_all('td')[6].text.strip()
    df.loc[idx, 'Goals Against (GA)'] = team.find_all('td')[7].text.strip()

df
```

```python
# Save to CSV
df.to_csv('hockey_teams_data.csv', index=False)
```
