```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Fetch the books information
res = requests.get('https://books.toscrape.com/')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# Create a DataFrame
df = pd.DataFrame(columns=['Name', 'Price', 'Rate'])
```

```python
# Parse the books data
for idx, book in enumerate(soup.find_all('article', 'product_pod')):
    df.loc[idx , 'Name'] = book.h3.a['title']
    df.loc[idx , 'Price'] = float(book.find_all('div')[1].p.text[2:])
    df.loc[idx , 'Rate'] = book.p['class'][1]

df
```

```python
# Save to CSV
df.to_csv('books_data.csv', index=False)
```
